```python
from data_clearning import Merge_ipv4_query

def username_IP_affinity(username):
    groupby_ipv4 = {ip:set(list(df.userName)) for ip,df in tb.groupby(["ipv4"])}
    user = set(list(tb[tb.ipv4== user_filter(username)[0]].userName))
    temp = {}
    temp_set_info = {}
    result = {"ipv4":[],"jaccard":[],"user":[],"intersection":[],"and_set":[]}
    for ip,username in groupby_ipv4.items():
        intersection = username & user
        and_set = username | user
        temp.update({ip:(len(intersection) / len(and_set))})
        temp_set_info.update({ip:[intersection,and_set]})
        #temp_user.update({ip:username})
    for iter_ in sorted(temp.items(), key = lambda kv:(kv[1], kv[0]), reverse=True ):
        if iter_[-1] > 0:
            result["jaccard"].append(iter_[-1])
            result["ipv4"].append(iter_[0])
            result["user"].append(groupby_ipv4[iter_[0]])
            intersection_and_set = temp_set_info[iter_[0]]
            intersection,and_set = intersection_and_set[0],intersection_and_set[-1]
            result["intersection"].append(intersection)
            result["and_set"].append(and_set)
    affinity = pd.DataFrame(result)
    ip_info = Merge_ipv4_query(result["ipv4"])
    tbmerge = pd.merge(affinity,ip_info, how="outer", on="ipv4")
    return tbmerge

Zhihui_IPV4 = username_IP_affinity("zhihui")
```

from jarvis_math import jaccard
def jaccard(A,B):
    '''
    雅卡尔相似度
    '''
    if isinstance(A,(list,tuple,np.ndarray)) and isinstance(B,(list,tuple,np.ndarray)):
        init_A,init_B = set(A),set(B)
    elif isinstance(A,set) and isinstance(B,set):
        init_A,init_B = A,B
    else:
        print("ERROR : Incorrect data type")
        return None
    intersection = init_A & init_B
    union = init_A | init_B
    return len(intersection) / len(union)

def jaccard_distance(A,B):
    '''
    雅卡尔距离
    '''
    if isinstance(A,(list,tuple,np.ndarray)) and isinstance(B,(list,tuple,np.ndarray)):
        init_A,init_B = set(A),set(B)
    elif isinstance(A,set) and isinstance(B,set):
        init_A,init_B = A,B
    else:
        print("ERROR : Incorrect data type")
        return None
    intersection = init_A & init_B
    union = init_A | init_B
    return (len(union) - len(intersection)) / len(union)


def cos(A,B):
    '''
    余弦相似度
    '''
    if len(A) != len(B):
        print("ERROR : The vectors have different lengths ")
        return None
    elif isinstance(A,(tuple,set)) and isinstance(B,(tuple,set)):
        init_A,init_B = np.array(list(A)),np.array(list(B))
    elif isinstance(A,(tuple,list)) and isinstance(B,(tuple,list)):
        init_A,init_B = np.array(A),np.array(B)
    elif isinstance(A,np.ndarray) and isinstance(B,np.ndarray):
        init_A,init_B = A,B
    else:
        print("ERROR : Incorrect data type")
        return None
    unit_vector = np.ones_like(init_A)
    return (init_A * init_B).dot(unit_vector)/(np.sqrt((init_A ** 2).dot(unit_vector)) * np.sqrt((init_B ** 2).dot(unit_vector)) )

def feature_dictionary_editor(iter_obj):
    '''
    特征字典编辑器
    '''
    drop_vector = [j for i in iter_obj for j in i ]
    unit_vector = {iter_:0 for iter_ in drop_vector}
    for iter_ in drop_vector:
        unit_vector[iter_] += 1
    return np.array(unit_vector)

feature_dictionary = feature_dictionary_editor(Zhihui_IPV4.user)

def feature_value_filter(A,B,return_all = False):
    """
    两个可迭代对象的计数余弦相似度
    """
    A,B = list(A),list(B)
    lenght_A,lenght_B = len(A),len(B)
    feature_dictionary = lambda : {iter_:0 for iter_ in (A+B)}
    if lenght_A > lenght_B:
        max_iter = zip(B,A+[None]*(lenght_A-lenght_B)) 
    else :
        max_iter = zip(B+[None]*(lenght_B-lenght_A),A)
    feature_dictionary_A,feature_dictionary_B = feature_dictionary(),feature_dictionary()
    for b,a in max_iter:
        feature_dictionary_B[b]+= 1
        feature_dictionary_A[a]+= 1
    # 过滤计数值
    filter_values = lambda x : [v for _,v in x.items()]
    

    cos_value = cos(filter_values(feature_dictionary_A),filter_values(feature_dictionary_B))
    if return_all == True: 
        return {"cos_value":cos_value,"feature_dictionary":feature_dictionary(),f"eature_dictionary_A":feature_dictionary_A,"feature_dictionary_B":feature_dictionary_B,"jaccard":jaccard(A,B)}
    elif return_all == "cos":
        return cos_value
    elif return_all == "jaccard":
        return jaccard(A,B)
    else:
        return cos_value,jaccard(A,B)

```python
from jarvis_math import feature_value_filter
from data_clearning import Merge_ipv4_query

def specific_user_infomation_table(username):
    '''
    包含某个字段的用户名列表
    '''
    usernames = tb.userName.apply(lambda x: x.lower())
    username = username.lower()
    return tb[np.array([username in name for name in usernames])]

def specific_user_infomation_tables(userNameList):
    '''
    包含某些字段的用户名列表
    '''
    if isinstance(userNameList,str):
        userNameList = [userNameList]
    elif isinstance(userNameList,(list,set,np.ndarray)):
        userNameList = userNameList
    return {name : specific_user_infomation_table(name) for name in userNameList}

def specific_ipv4_infomation_table(ipv4):
    '''
    包含特定ip的信息列表
    '''
    ipv4_info_table = tb[tb.ipv4 == ipv4]
    return ipv4_info_table

#ip_with_similar_users(["zhanglin","zhihui","zhangjia"])

def specific_user_association_information_tables(username):
    '''
    用户名关联的列表
    '''
    result = {"ipv4_1":[],"ipv4":[],"cos":[],"jaccard":[],"check_name_lenght":[],"return_name_lenght":[],"same_ip_usernames":[],"associated_with_ip_usernames":[]}
    groupby_ipv4 = {ipv4:df for ipv4,df in tb.groupby(["ipv4"])}
    associated_ip = set(specific_user_infomation_table(username).ipv4)
    associated_ips_usernames = {ipv4:specific_ipv4_infomation_table(ipv4) for ipv4 in associated_ip}
    for ipv4_1,table in associated_ips_usernames.items():
        user_names  = list(table.userName)
        for ipv4_2,ipv4_on_this_table in groupby_ipv4.items():
            ipv4_2_user_names = list(ipv4_on_this_table.userName)
            feature_value = feature_value_filter(user_names,ipv4_2_user_names)
            #print(feature_value)
            jaccard_value,cos_value,check_name_lenght,return_name_lenght,same_ip_usernames,associated_with_ip_usernames = feature_value["cos"],feature_value["jaccard"],feature_value["lenght_A"],feature_value["lenght_B"],feature_value["feature_dictionary_A"],feature_value["feature_dictionary_B"]
            if (jaccard_value != 0) and (cos_value > 0):
                result["cos"].append(jaccard_value)
                result["jaccard"].append(cos_value)
                result["ipv4_1"].append(ipv4_1)
                result["ipv4"].append(ipv4_2)
                result["check_name_lenght"].append(check_name_lenght)
                result["return_name_lenght"].append(return_name_lenght)
                result["same_ip_usernames"].append(same_ip_usernames)
                result["associated_with_ip_usernames"].append(associated_with_ip_usernames)


    ipv4s_2_info_table = Merge_ipv4_query(set(result["ipv4"]))
    
    df = pd.merge(pd.DataFrame(result),ipv4s_2_info_table,how='outer',on='ipv4') 
    return df


#zhangjia_table = specific_user_association_information_tables("zhangjia")#,specific_ipv4_infomation_table("45.248.189.3")
#zhangjia_table.to_csv("zhangjia.csv")
```
